#**LIMPIEZA**游빛

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ','_')
```

#**Eliminar filas o columnas completamente vacias**游뛅

```python
df.dropna(axis = 0, how = "all", inplace = True ) #fila = 0
df.dropna(axis = 1, how = "all", inplace = True ) #columna = 1
```
